```python
plt.figure(figsize=(10,5))

sns.barplot(x=df['Region'],y=df[' Estimated Labour Participation Rate (%)'])
plt.title('Region wise Estimated Unemployment Rate')

plt.xticks(rotation=50)
plt.style.use('classic')
plt.plot()
```
